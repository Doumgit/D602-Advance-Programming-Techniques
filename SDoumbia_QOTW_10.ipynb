## 1. What is matplottlib and seaborn?  When would you choose one over the other?

- **What is matplottlib and seaborn?**

  - **Matplotlib**: This is a powerful plotting library for Python and is the foundation for many other plotting libraries in Python. It provides a wide variety of plotting functions and options, allowing  to create virtually any kind of plot. Matplotlib is highly customizable, but this level of customization sometimes requires writing more code. It's well-suited for creating basic plots as well as highly customized visualizations.

  - **Seaborn**: Seaborn is built on top of Matplotlib and provides a higher-level interface for creating attractive and informative statistical graphics. It simplifies the process of creating complex visualizations by providing easy-to-use functions for common statistical plots, such as scatter plots, bar plots, box plots, violin plots, and more. Seaborn also offers built-in themes and color palettes to make your plots visually appealing with minimal effort. It's particularly useful for exploratory data analysis and when we need to quickly create visually appealing plots.

- **When would you choose one over the other?**

  - **Use Matplotlib** when we need highly customized plots and fine control over every aspect of the visualization. If we have specific requirements that go beyond what Seaborn provides, Matplotlib gives us the flexibility to create the exact plot you want.

  - **Use Seaborn** when we want to quickly create attractive, informative visualizations without writing a lot of code. It's great for exploratory data analysis and when we need to produce publication-quality plots with minimal effort. Seaborn's high-level functions and built-in themes make it easy to create visually appealing plots with just a few lines of code.

## 3. Give an example of either a matplotlib or seaborn graphic (give code).  You may also reference an informative article.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Creating a bar plot using seaborn
sns.barplot(x="class", y="survived", data=titanic, errorbar=None)

# Adding labels and title
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Average Survival Rate by Passenger Class")

# Showing the plot
plt.show()
```

**What is the data-ratio?**

It is a concept which refers to the proportion of ink (or pixels in digital visualization) used to represent the actual data compared to the total amount of ink used in the visualization.
In essence, maximizing the data-ink ratio involves minimizing unnecessary ink or elements in a visualization that do not directly contribute to conveying the underlying data. By reducing unnecessary clutter and maximizing the ink dedicated to representing the data, visualizations become clearer, more concise, and more effective in communicating insights.

**How can you use python libraries such as matplotlib, seaborn, plotly, etc. to incorporate this data-ink ratio in your visualizations?**

We can follow these guidelines to incorporate it in our visualization:

  - **Simplify  plots:** Remove unnecessary elements such as borders, gridlines, and background colors that do not directly convey information about the data.
  - **Use minimalistic design:** Opt for clean and simple design choices, such as using a limited color palette, avoiding unnecessary decorations, and minimizing the use of 3D effects or embellishments.
  - **Choose appropriate plot types:** Select the most appropriate plot type that effectively communicates the patterns and relationships present in our data without adding unnecessary complexity. For example, if we have time-series data, a simple line plot might be more effective than a 3D surface plot.
  - **Reduce redundant information:** Avoid redundant labeling or annotations that do not add value to the visualization. Use concise and informative titles, axis labels, and annotations to convey key insights without cluttering the plot.
  -**Maximize data-ink:** Ensure that the majority of ink or pixels in our visualization are dedicated to representing the actual data rather than non-data elements. Eliminate chart junk and non-data ink to maximize the data-ink ratio and improve the clarity of the visualization.
