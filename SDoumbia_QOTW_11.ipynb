## 1. How do you plot a histogram in Seaborn?  

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset('penguins')

# Create a histogram
sns.histplot(data=penguins, x='body_mass_g', bins=20, kde=True,color='blue')

# Adding labels and title
plt.title('Distribution of Penguin Body Mass')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## 2. Plot a histogram with NAs dropped.

```python
# Drop NA values from the 'body_mass_g' column
penguins = penguins.dropna(subset=['body_mass_g'])

# Create a histogram with NAs dropped
sns.histplot(data=penguins, x='body_mass_g', bins=20, kde=True, color='blue')

# Adding labels and title
plt.title('Distribution of Penguin Body Mass - Excluding NAs')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## 3. How do you set the color for a histogram?

In the histogram above, the color parameter is used within **sns.histplot()** to set the color of the histogram. We can replace **'blue'** with any color name (like **'red'**, **'green'** etc.), or we can use **hexadecimal color codes (like '#FF5733')**.

## 4.  What type of plot would allow you to compare two continuous features?  Give an example of code.

To compare two continuous features, **scatter plots** are particularly useful. They allow us to visualize the relationship between two variables, helping to spot correlations, trends, or potential outliers. In Seaborn, we can create a scatter plot using the **'.scatterplot()'** function.

```python
# Drop NA values from relevant columns
penguins = penguins.dropna(subset=['flipper_length_mm', 'body_mass_g'])

# Create a scatter plot
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g')

# Adding labels and title
plt.title('Relationship between Flipper Length and Body Mass in Penguins')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')

# Show the plot
plt.show()
```

## 5. Give example of a correlation plot.

```python
# Drop NA values to ensure accurate correlation computation
penguins = penguins.dropna()

# Select only numerical columns for the correlation matrix
numerical_penguins = penguins.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr = numerical_penguins.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

# Adding title
plt.title('Correlation Matrix for Penguin Dataset')

# Show the plot
plt.show()
```

## 6. Change the figure size of your plot(s).

**Correlation PLot**

```python
penguins = penguins.dropna()
numerical_penguins = penguins.select_dtypes(['float64', 'int64'])
corr = numerical_penguins.corr()

plt.figure(figsize=(10, 8))  # Setting the figure size
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Penguin Dataset')
plt.show()
```
